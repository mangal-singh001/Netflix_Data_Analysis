Q 2.) In which year highest number of the TV shows & Movies were related ? Show with bar graph

```python
data.dtypes
```

```python
# Make a new column of the right date format
data['Date_N'] =  pd.to_datetime(data['Release_Date'].str.strip(),
errors='coerce')
print(data.head())
```

```python
print(data.dtypes)
```

```python
dc = data['Date_N'].dt.year.value_counts()
```

```python
# In 2019 most of the Movies and TV Shows are released
bar = data['Date_N'].dt.year.value_counts().sort_index()
plt.figure(figsize=(12,6))
sns.barplot(x=bar.index,y=bar.values,palette='viridis')
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.title("Number of Movies and TV Shows Released each Year")
plt.tight_layout()
plt.show()
```

```python
data['Date_N'].dt.year.value_counts().plot(kind='bar')
```
